### Create a hierarchical index

```python
# create hiearchical index based off the Country column
candy_df.set_index(['Internal ID', 'Q4: COUNTRY'], inplace=True)

print(candy_df)
```

### Pivot the data

```python
# use 'melt' to pivot the data
melted_df = candy_df.melt(id_vars = 'Internal ID', value_vars = ['Q1: GOING OUT?', 'Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY', 'Q5: STATE, PROVINCE, COUNTY, ETC'])
pivot_candy_df = melted_df.pivot(index = 'Internal ID', columns = 'variable', values = 'value')

pivot_candy_df.head()
```

### Grouping with Dicts/Series

```python
# group the dataset by country
grouped_candy_df = candy_df.groupby('Q4: COUNTRY')

print(grouped_candy_df)
```

```python
# use a 'for' loop to iterate over the data and print the results
for group, data in grouped_candy_df:
    print(f"Group: {group}")
    print(data)
    print()
```

### Cross Tabs

```python
# create cross-tabs based off of Gender and Age
cross_tab = pd.crosstab(index = candy_df['Q2: GENDER'], columns = candy_df['Q3: AGE'])

cross_tab.head()
```

### Import Excel file with date and time data

```python
# read the Excel file into a DataFrame
candy_2015survey_df = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx')
candy_2015survey_df.head()
```

### Convert between string and date time

```python
candy_2015survey_df['Datetime'] = pd.to_datetime(candy_2015survey_df['Timestamp'], format='%m/%d/%Y %H:%M:%S')
print(candy_2015survey_df['Datetime'])
```

### Generate date range

```python
date_range = pd.date_range(start = candy_2015survey_df['Timestamp'].min(), end = candy_2015survey_df['Timestamp'].max(), freq='D')

print(date_range)
```
